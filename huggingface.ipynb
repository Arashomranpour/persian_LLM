```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS
from langchain.llms import CTransformers
from langchain.chains import ConversationalRetrievalChain
```

```python
df1=df1.replace({"\u200d":" "})
df1=df1.replace({"\u200c":" "})
df1.to_csv("file.csv")
```

```python
data=df1['context'].tolist()
```

```python
# print(df1)
print(len(df1))
```

```python
df1.head(2)
```

```python
text_splitter=RecursiveCharacterTextSplitter(chunk_size=2000,chunk_overlap=200)
text_chunks=text_splitter.create_documents(data)
```

```python
len(text_chunks)
```

```python
embedding=HuggingFaceEmbeddings(model_name="BAAI/bge-m3")
```

```python
from langchain.vectorstores import FAISS
from langchain.llms import CTransformers
from langchain.chains import ConversationalRetrievalChain
from langchain.memory import ConversationBufferMemory
```

```python
docsearch=FAISS.from_documents(text_chunks,embedding)
```

```python
docsearch.save_local("./")
```

```python
query="کتاب مقدس دین اسلام چیست؟"
docs=docsearch.similarity_search(query=query,k=1)
print(str(docs[0]).replace("\u200c"," "))
```

ChromaDB

```python
from langchain.vectorstores.chroma import Chroma
persist_directory="db"
db = Chroma.from_documents(text_chunks, embedding, persist_directory=persist_directory)

db.persist()
vect=Chroma(persist_directory=persist_directory,embedding_function=embedding)
retriver=vect.as_retriever( search_kwargs={"k":1})
```

```python
res=retriver.get_relevant_documents("طوایف عرب خوزستان همگی از چه دسته‌ ای بودند؟")
print(str(res[0]).replace("\u200c"," "))
```

```python
res=retriver.get_relevant_documents("در بعضی از گویش‌های رایج در استان خوزستان «بختیاری» برای ادای حرف «خ» از چه حرفی بهره می‌گیرد؟")
print(str(res[0]).replace("\u200c"," "))
```
